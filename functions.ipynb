## Построние графиков

Matplotlib - это библиотека на языке программирования Руthоn для создания графиков и визуализации данных. Она предоставляет широкий спектр возможностей для создания различных видов графиков, включая:
- Линейные графики: простые многострочные, точечные шаговые, шестиугольные и т.д.
- Гистограммы: вертикальные, горизонтальные, наклонные, с различными стилями заполнения и т.д.- Круговые диаграммы: стандартные, вложенные, секторные и т.д.
- Графики рассеяния: одномерные, двумерные, многомерные, с настройкой размера и цвета точек и т.д.
- Контурные и поверхтностные графики: линии уровня, изолинии, тепловые карты, трехмерные графики и т.д.
- Диаграммы рассеивания и ящики с усами: создание диаграмм рассеивания, ящиков с усами, настраиваемых маркеров и цветов.
- Графики временных рядов: создание линейных графиков временных рядов с настраиваемыми осями и метками времени.
- Диаграммы ошибок: шрафики ошибок, погрешностей и отклонений, графики с асимметричными ошибками и т.д.
- Анимации: создание анимированных графиков, включая изменение позиции, размера, цвета и т.д.

Это далеко не полный список возможностей, которые предоставляет библиотека Matplotlib, Она также имеет широкий спектр настроек и параметров для настройки внешнего вида графиков и диаграмм, таких как цвета, шрифты, легенды, подписи и т.д.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### $f(x) = a \cdot x + b$

```python
# определение значений x
x = np.linspace(-10, 10, 100)

# определение параметров a и b
a = 2
b = 1

# вычисление значений y
y = a * x + b

# построение графика
plt.plot(x, y, color="blue", linestyle="-", linewidth=2)

# настройка осей координат и заголовка
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.title("График функции y = a*x+b")
plt.axis("equal")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(y=0, color='black', linewidth=0.5, linestyle='-')  # ось X
plt.axvline(x=0, color='black', linewidth=0.5, linestyle='-')  # ось Y

# отображение графика
plt.show()
```

### $f(x) = x^2$

```python
# определение значений x
x = np.linspace(-10, 10, 100)

# вычисление значений y
y = x ** 2

# построение графика
plt.plot(x, y, color="blue", linestyle="-", linewidth=2)

# настройка осей координат и заголовка
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.title("График функции y = x**2")
plt.axis("equal")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(y=0, color='black', linewidth=0.5, linestyle='-')  # ось X
plt.axvline(x=0, color='black', linewidth=0.5, linestyle='-')  # ось Y

# отображение графика
plt.show()
```

### $f(x) = \large\frac{1}{x}$

```python
# определение значений x, за исключением точки x=0
x = np.linspace(-10, 10, 200)
x1 = x[:100]
x2 = x[101:]

# вычисление значений y, за исключением точки x=0
y1 = 1/x1
y2 = 1/x2

# построение графика
plt.plot(x1, y1, color="blue", linestyle="-", linewidth=2)
plt.plot(x2, y2, color="blue", linestyle="-", linewidth=2)

# настройка осей координат и заголовка
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.title("График функции y = 1/x")
plt.axis("equal")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(y=0, color='black', linewidth=0.5, linestyle='-')  # ось X
plt.axvline(x=0, color='black', linewidth=0.5, linestyle='-')  # ось Y

# отображение графика
plt.show()
```
